# Задание 1 

Напишите функцию, которая принимает на вход строку и проверяет, является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2–3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:

car_id = 'А222BС96'

Результат: Номер А222BС валиден. Регион: 96.


car_id = 'АБ22ВВ193'

Результат: Номер не валиден.


```python
import re
```

```python
# функция определния валидности номера
def validate_car_number(car_id):
    # Проверка транспортного номера 1 буква, 3 цифры, 2 буквы, 2-3 цифры
    pattern = r'^([АВЕКМНОРСТУХ])(\d{3})([АВЕКМНОРСТУХ]{2})(\d{2,3})$'
    
    match = re.match(pattern, car_id)
    
    if match:
        # Извлекаем группы: буква, цифры, буквы, регион
        letter = match.group(1)
        digits = match.group(2)
        letters = match.group(3)
        region = match.group(4)
        
        # Формируем номер
        number = letter + digits + letters
        
        return f"Номер {number} валиден. Регион: {region}."
    else:
        return f"Номер {car_id} не валиден."
```

```python
# проверка
# основная проверка по примеру
print("=== Основные примеры ===")
print(validate_car_number('А222ВС96'))    # Номер А222BС валиден. Регион: 96.
print(validate_car_number('АБ22ВВ193'))   # Номер не валиден.

print("\n=== Дополнительные валидные номера ===")
print(validate_car_number('К123ТУ777'))   # Номер К123ТУ валиден. Регион: 777.
print(validate_car_number('М555ОО123'))   # Номер М555ОО валиден. Регион: 123.
print(validate_car_number('В001АВ02'))    # Номер В001АВ валиден. Регион: 02.
print(validate_car_number('Е777КХ177'))   # Номер Е777КХ валиден. Регион: 177.
print(validate_car_number('Х999РТ99'))    # Номер Х999РТ валиден. Регион: 99.

print("\n=== Дополнительные невалидные номера ===")
print(validate_car_number('А1234ВС77'))   # Невалиден - 4 цифры в номере
print(validate_car_number('А12ВВ777'))    # Невалиден - 2 цифры в номере
print(validate_car_number('Я123АВ77'))    # Невалиден - буква Я не разрешена
print(validate_car_number('А123В1'))      # Невалиден - слишком короткий
print(validate_car_number('А123ВВ1'))     # Невалиден - регион из 1 цифры
print(validate_car_number('А123ВВ1234'))  # Невалиден - регион из 4 цифр
print(validate_car_number('123АВВ77'))    # Невалиден - начинается с цифры
print(validate_car_number('БИ123СТ45'))   # Невалиден - 2 буквы в начале
```

Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.


Пример работы программы:
some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'


Результат: Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.


```python
def remove_duplicate_words(text):
    # Регулярное выражение для поиска последовательных повторяющихся слов
    pattern = r'\b(\w+)\s+\1\b'
    
    # Повторяем замену пока есть изменения
    previous_text = ""
    while previous_text != text:
        previous_text = text
        text = re.sub(pattern, r'\1', text)
        
    result = re.sub(pattern, r'\1', text, flags=re.IGNORECASE)
    return result
```

```python
# Пример использования
some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'
```

```python
result = remove_duplicate_words(some_string)
print("Исходная строка:")
print(some_string)
print("\nРезультат:")
print(result)
```
